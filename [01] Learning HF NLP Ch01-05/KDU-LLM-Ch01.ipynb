```python
# Importing necessary libraries
from transformers import pipeline
```

```python
# Use 'sentiment-analysis' to predict sentiment of a sentence
classifier = pipeline("sentiment-analysis")
```

```python
# One sentence example
print(classifier("Sad"))
print(classifier("Happy"))
print(classifier("Gloomy"))

# Multiple sentences example
print(classifier("I love doing ice skating!"))
print(classifier("I don't want to watch television anymore"))
```

```python
# Use 'zero-shot-classification' to classify a label not just a sentiment
classifier = pipeline("zero-shot-classification")
```

```python
print(
    classifier(
    "Hmmm, I think bill gates had something to do with open AI drama",
    candidate_labels = ["Hoax", "Business", "Technology", "News"]
    )
)

print(
    classifier(
    "Indonesia's currency will inflate 5 times in the next month",
    candidate_labels = ["Economy", "Business", "Technology", "News"]
    )
)
```

```python
# Use 'text-generation' to generate text with GPT-2 as a default model
generator = pipeline("text-generation")
```

```python
generator("Indonesia is")
```

```python
# Use 'mask filling' to fill the mask in a sentence
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=1)
```

```python
# Use 'ner' to recognize words in a sentences (Named Entity Recognition)
ner = pipeline("ner", grouped_entities=True)
ner("My name is Tirta and I work at Kalbe in Jakarta.")
```

```python
# Use 'summarization' to summarize a text
summarizer = pipeline("summarization")
summarizer(
    """
    On January 11, Navas arrived at the park, bought his ticket, and rented a basic diamond hunting kit, according to the news release.

“I got to the park around nine o’clock and started to dig,” Navas said in the release. “That is back-breaking work, so by the afternoon I was mainly looking on top of the ground for anything that stood out.”

Lucky for Navas, the park had received more than an inch of rain a few days before he arrived, so it was wet and muddy, the release said.

“As rain falls on the field, it washes away the dirt and uncovers heavy rocks, minerals and diamonds near the surface,” Assistant Park Superintendent Waymon Cox explained.

Many of the park’s biggest diamonds are found on the surface, Cox said, and the park periodically plows the 37.5 acre search area to loosen the soil and to promote natural erosion.

Eventually, Navas emerged at the park’s Diamond Discovery Center with his findings. There, he was told he had found a 7.46-carat brown diamond.
"""
)
```
